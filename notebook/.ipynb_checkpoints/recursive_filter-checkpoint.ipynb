```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## make measurements
dt = 0.2
t  = np.arange(0, 10, dt)

xm = np.zeros([len(t)], dtype=np.float32)

# get volt function
def get_volt(k=1):
    w = 0. + 4.*k*np.random.rand()
    z = 14.4 + w
    
    return z

for k in range(len(t)):
    xm[k] = get_volt(k)
```

```python
# average filter
class recursive_filter(object):
    def __init__(self):
        self.prev_avg = 0;
        self.k = 0
        
    def avg_filter(self, x):
        self.k = self.k + 1
        
        alpha = (self.k-1) / self.k
        avg = alpha*self.prev_avg + (1-alpha)*x

        self.prev_avg = avg

        return avg
    
    def moving_avg(self, x):
        if (self.k == 0):
            self.prev_avg = x
            self.xbuf = np.ones(self.n+1)*x
            
        
        for i in range(self.n):
            self.xbuf[i] = self.xbuf[i+1]
            
        self.xbuf[self.n] = x
        
        avg = self.prev_avg + (x - self.xbuf[0]) / (float)(self.n)
        self.prev_avg = avg
        
        self.k = self.k + 1
        
        return avg
```

```python
## test
f = recursive_filter()

avg_rec = np.zeros([len(t)])

for k in range(len(t)):
    avg_rec[k] = f.avg_filter(xm[k])
```

```python
plt.hold('on')
plt.plot(xm)
plt.plot(avg_rec)
plt.title('recursive average filter')
plt.legend(['measurements','average filter'])
```

Moving average:
$$
\bar x_k = \frac{x_{k-n+1} + x_{k-n+2} + \cdots + x_k}{n}
$$
To derive a recursive formula, see the previous step
$$
\bar x_{k-1} = \frac{x_{k-n} + x_{k-n+1} + \cdots + x_{k-1}}{n}
$$
Substract:
$$
\bar x_k - \bar x_{k-1} = \frac{x_k - x_{k-n}}{n}
$$

```python
# moving average
class recursive_filter(object):
    def __init__(self):
        self.prev_avg = 0
        self.k = 0
        self.n = 10
        
    def moving_avg(self, x):
        
        if (self.k == 0):
            self.prev_avg = x
            self.xbuf = np.ones(self.n+1)*x
            
        
        for i in range(self.n):
            self.xbuf[i] = self.xbuf[i+1]
            
        self.xbuf[self.n] = x
        
        avg = self.prev_avg + (x - self.xbuf[0]) / (float)(self.n)
        self.prev_avg = avg
        
        self.k = self.k + 1
        
        return avg
```

```python
## test
# dt = 0.2
# t  = np.arange(0, 10, dt)
f  = recursive_filter()

avg_moving = np.zeros([len(t)])

for k in range(len(t)):
    avg_moving[k] = f.moving_avg(xm[k])
```

```python
plt.hold('on')
plt.plot(xm)
plt.plot(avg_rec)
plt.plot(avg_moving)
plt.title('recursive average filter')
plt.legend(['measurements','average recursive','moving average'])
```
